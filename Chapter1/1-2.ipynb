使用設定檔建立資料庫連線

```python
import pymysql
from configparser import ConfigParser

# 讀取 config.ini 檔案取得資料庫連線資訊，並可以取用該檔案的資料
config = ConfigParser()
config.read('config.ini')

# 建立資料庫連線
connection = pymysql.connect(
    host=config['DB']['host'],
    user=config['DB']['user'],
    password=config['DB']['password'],
    port=int(config['DB']['port']), # 用 ConfigParser 讀取出來的資料，會被預設為 str 型別，但是 port 必須是 int 型別，所以這邊要強制轉型
    cursorclass=pymysql.cursors.DictCursor,
)

with connection.cursor() as cursor:
    query = """
        SHOW DATABASES;
    """
    cursor.execute(query)
    result = cursor.fetchall()
print(result)
```
