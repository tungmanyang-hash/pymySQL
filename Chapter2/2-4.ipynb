```python
db_connect = {
    "host": config.get('DB', 'host'),
    "user": config.get('DB', 'user'),
    "password": config.get('DB', 'password'),
    "port": config.getint('DB', 'port'),
    "cursorclass": pymysql.cursors.DictCursor,
}
 
# 特例 -> 因為 CREATE DATABASE 不能帶入字串
def create_database(database):
    connection = pymysql.connect(**db_connect)
    with connection.cursor() as cursor:
        sql = f"""
            CREATE DATABASE IF NOT EXISTS {database};
        """
        # 執行建立的 SQL 語句
        cursor.execute(sql)
        # 執行查看資料庫
        cursor.execute("SHOW DATABASES;")
        dbs = cursor.fetchall()
 
        print(dbs)
```

```python
# keyword arguments 是一個 dict，可以應用到把 connection 縮減排數

# 參數很多 function 都一樣，要怎麼縮減 -> **kwarge

name = {'b':'Tim', 'age':20}
info = {'a':20}

# 也可以用在 dict 的合併 -> 在想帶入的 dict 內逗號，後方加入 ** 後直接輸入想加入的 dict 的變數(dict 等於左邊的 dict 名字) 

info = {'a':20, **name}
print(info)

# 甚至加入後修改某個 key 的 value：在想帶入的 dict 內逗號，後方加入 ** 後直接輸入想加入的 dict 的變數，再加一個逗號，輸入 key: 想更改的 value，想更改的 value 就會重新被賦值
info = {'a':20, **name, 'age':15}
print(info)
```
